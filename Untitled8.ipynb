```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
import logging
from typing import Optional, Dict
from datetime import datetime

# --- Configuration ---
DATA_FILE_NAME = "OpenSea_NFT_Sales_2019_2021.csv"
PLOTS_OUTPUT_DIR = "nft_plots"
LOG_FILE = "nft_analysis.log"

# Setup logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler(LOG_FILE),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)
```

```python
# --- Helper Functions ---

def load_and_initial_inspect(file_path: str) -> Optional[pd.DataFrame]:
    logger.info(f"Attempting to load dataset '{file_path}'")
    if not os.path.exists(file_path):
        logger.error(f"File '{file_path}' not found.")
        return None

    try:
        df = pd.read_csv(file_path, low_memory=False, dtype={
            'asset.collection.short_description': str,
            'total_price': str
        }, on_bad_lines='skip')
        logger.info(f"Dataset '{file_path}' loaded successfully! Rows: {len(df)}")

        logger.info("First 5 rows of the dataset:")
        print(df.head())

        logger.info("Overall dataset information:")
        print(df.info())

        if 'Category' in df.columns:
            logger.info("Unique values of 'Category' column:")
            print(df['Category'].dropna().unique())
        else:
            logger.warning("'Category' column not found. Available columns:")
            print(df.columns)

        return df

    except Exception as e:
        logger.error(f"Error loading dataset: {e}")
        return None

def categorize_nft(collection_name: str) -> str:
    collection_name = str(collection_name).lower()
    if any(keyword in collection_name for keyword in ['crypto', 'punk', 'ape']):
        return 'Collectibles'
    elif any(keyword in collection_name for keyword in ['land', 'decentraland', 'sandbox']):
        return 'Virtual Worlds'
    elif 'art' in collection_name or 'gallery' in collection_name:
        return 'Art'
    elif 'music' in collection_name:
        return 'Music'
    elif 'photo' in collection_name or 'photography' in collection_name:
        return 'Photography'
    else:
        return 'Uncategorized'
```

```python
# --- Helper Functions (Continued) ---

def preprocess_nft_data(df: pd.DataFrame) -> Optional[pd.DataFrame]:
    if df is None or df.empty:
        logger.error("DataFrame is empty or None, preprocessing skipped.")
        return None

    logger.info("Starting data preprocessing")
    df_processed = df.copy()
    if 'total_pric' in df_processed.columns:
        df_processed.rename(columns={'total_pric': 'total_price'}, inplace=True)
        logger.info("Renamed 'total_pric' to 'total_price'.")

    try:
        df_processed['total_price_ether'] = pd.to_numeric(df_processed['total_price'], errors='coerce') / 1e18
        invalid_rows = df_processed['total_price_ether'].isna().sum()
        df_processed.dropna(subset=['total_price_ether'], inplace=True)
        logger.info(f"Converted 'total_price' to 'total_price_ether'. Dropped {invalid_rows} invalid rows.")
    except Exception as e:
        logger.error(f"Error converting 'total_price': {e}")
        return None

    df_processed['sales_datetime'] = pd.to_datetime(df_processed['sales_datetime'], errors='coerce')
    invalid_dates = df_processed['sales_datetime'].isna().sum()
    df_processed.dropna(subset=['sales_datetime'], inplace=True)
    logger.info(f"Converted 'sales_datetime' to datetime. Dropped {invalid_dates} invalid rows.")

    df_processed['sale_year'] = df_processed['sales_datetime'].dt.year
    df_processed['sale_month'] = df_processed['sales_datetime'].dt.month
    df_processed['sale_day_of_week'] = df_processed['sales_datetime'].dt.day_name()
    logger.info("Extracted year, month, and day-of-week features.")

    # Address FutureWarning by not using inplace=True directly on chained assignments
    df_processed['asset.name'] = df_processed['asset.name'].fillna('Unknown')
    df_processed['asset.collection.name'] = df_processed['asset.collection.name'].fillna('Uncategorized')
    df_processed['asset.collection.short_description'] = df_processed['asset.collection.short_description'].fillna('No Description')
    logger.info("Handled missing values in 'asset.name', 'asset.collection.name', and 'asset.collection.short_description'.")

    df_processed['Category'] = df_processed['asset.collection.name'].apply(categorize_nft)
    logger.info("Created 'Category' column based on 'asset.collection.name'.")

    logger.info("DataFrame info after preprocessing:")
    print(df_processed.info())
    logger.info("First 5 rows after preprocessing:")
    print(df_processed.head())

    return df_processed

def perform_eda(df: pd.DataFrame) -> Dict:
    if df is None or df.empty:
        logger.error("DataFrame is empty or None, EDA skipped.")
        return {}

    logger.info("Starting exploratory data analysis")
    eda_results = {}

    eda_results['sales_by_category'] = df['Category'].value_counts()
    eda_results['revenue_by_category'] = df.groupby('Category')['total_price_ether'].sum().sort_values(ascending=False)
    eda_results['avg_price_by_category'] = df.groupby('Category')['total_price_ether'].mean().sort_values(ascending=False)
    eda_results['sales_by_year'] = df.groupby(['sale_year', 'Category']).size().unstack(fill_value=0)
    eda_results['revenue_by_year'] = df.groupby(['sale_year', 'Category'])['total_price_ether'].sum().unstack(fill_value=0)

    logger.info("Completed EDA")
    return eda_results
```

```python
# --- Helper Functions (Continued) ---

def visualize_nft_data(df: pd.DataFrame, eda_results: Dict) -> None:
    if df is None or df.empty or not eda_results:
        logger.error("DataFrame or EDA results are empty, visualization skipped.")
        return

    logger.info("Starting data visualization")
    os.makedirs(PLOTS_OUTPUT_DIR, exist_ok=True)
    sns.set(style="whitegrid", palette="viridis")

    # 1. Number of NFT Sales by Category (Plotly)
    # Plotly figures generate HTML files; they won't show directly in the cell output
    # like Matplotlib plots, but the HTML file will be created.
    fig1 = px.bar(
        x=eda_results['sales_by_category'].index,
        y=eda_results['sales_by_category'].values,
        labels={'x': 'Category', 'y': 'Number of Sales'},
        title='Number of NFT Sales by Category (2019-2021)',
        color=eda_results['sales_by_category'].index
    )
    fig1.update_layout(xaxis_title='Category', yaxis_title='Number of Sales', xaxis_tickangle=45, showlegend=False)
    fig1.write_html(os.path.join(PLOTS_OUTPUT_DIR, 'sales_by_category.html'))
    logger.info("Saved sales by category plot (Plotly). You can view it by downloading 'sales_by_category.html'.")

    # 2. Total NFT Revenue (in Ether) by Category
    plt.figure(figsize=(12, 7))
    sns.barplot(x=eda_results['revenue_by_category'].index, y=eda_results['revenue_by_category'].values)
    plt.title('Total NFT Revenue (in Ether) by Category')
    plt.xlabel('Category')
    plt.ylabel('Total Revenue (Ether)')
    plt.xticks(rotation=60, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'revenue_by_category.png'))
    plt.show() # Display the plot after saving
    logger.info("Saved revenue by category plot.")

    # 3. Average NFT Sale Price (in Ether) by Category
    plt.figure(figsize=(12, 7))
    sns.barplot(x=eda_results['avg_price_by_category'].index, y=eda_results['avg_price_by_category'].values)
    plt.title('Average NFT Sale Price (in Ether) by Category')
    plt.xlabel('Category')
    plt.ylabel('Average Price (Ether)')
    plt.xticks(rotation=60, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'avg_price_by_category.png'))
    plt.show() # Display the plot after saving
    logger.info("Saved average price by category plot.")

    # 4. Annual Trend of NFT Sales by Category
    plt.figure(figsize=(14, 8))
    eda_results['sales_by_year'].plot(kind='line', marker='o')
    plt.title('Annual Trend of NFT Sales by Category')
    plt.xlabel('Year')
    plt.ylabel('Number of Sales')
    plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'annual_sales_trend.png'))
    plt.show() # Display the plot after saving
    logger.info("Saved annual sales trend plot.")

    # 5. Annual Trend of NFT Revenue (in Ether) by Category
    plt.figure(figsize=(14, 8))
    eda_results['revenue_by_year'].plot(kind='line', marker='o')
    plt.title('Annual Trend of NFT Revenue (in Ether) by Category')
    plt.xlabel('Year')
    plt.ylabel('Total Revenue (Ether)')
    plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'annual_revenue_trend.png'))
    plt.show() # Display the plot after saving
    logger.info("Saved annual revenue trend plot.")

    # New plots from previous suggestions:
    # 6. Monthly Trend of NFT Sales
    df_monthly_sales = df.set_index('sales_datetime').resample('M').size()
    plt.figure(figsize=(14, 7))
    df_monthly_sales.plot(kind='line', marker='o', color='purple')
    plt.title('Monthly Trend of NFT Sales (2019-2021)')
    plt.xlabel('Date')
    plt.ylabel('Number of Sales')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'monthly_sales_trend.png'))
    plt.show()
    logger.info("Saved monthly sales trend plot.")

    # 7. Distribution of NFT Sale Prices (in Ether)
    plt.figure(figsize=(10, 6))
    sns.histplot(df['total_price_ether'], bins=50, kde=True, color='teal')
    plt.title('Distribution of NFT Sale Prices (in Ether)')
    plt.xlabel('Sale Price (Ether)')
    plt.ylabel('Frequency')
    # plt.xscale('log') # Uncomment if prices vary widely and need log scale
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'price_distribution.png'))
    plt.show()
    logger.info("Saved price distribution plot.")

    # 8. NFT Sales by Day of Week
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='sale_day_of_week', order=day_order, palette='crest')
    plt.title('NFT Sales by Day of Week')
    plt.xlabel('Day of Week')
    plt.ylabel('Number of Sales')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'sales_by_day_of_week.png'))
    plt.show()
    logger.info("Saved sales by day of week plot.")

    # 9. Top 10 NFT Collections by Revenue
    # Ensure 'asset.collection.name' is filled for this plot to be meaningful
    top_collections_revenue = df.groupby('asset.collection.name')['total_price_ether'].sum().nlargest(10)
    plt.figure(figsize=(12, 8))
    sns.barplot(x=top_collections_revenue.values, y=top_collections_revenue.index, palette='rocket')
    plt.title('Top 10 NFT Collections by Revenue (in Ether)')
    plt.xlabel('Total Revenue (Ether)')
    plt.ylabel('Collection Name')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_OUTPUT_DIR, 'top_10_collections_revenue.png'))
    plt.show()
    logger.info("Saved top 10 collections by revenue plot.")

    logger.info("Visualization complete. Plots saved in 'nft_plots' folder.")

def save_results(eda_results: Dict) -> None:
    os.makedirs(PLOTS_OUTPUT_DIR, exist_ok=True)
    for key, value in eda_results.items():
        if isinstance(value, pd.Series):
            value.to_csv(os.path.join(PLOTS_OUTPUT_DIR, f"{key}.csv"))
        elif isinstance(value, pd.DataFrame):
            value.to_csv(os.path.join(PLOTS_OUTPUT_DIR, f"{key}.csv"))
        logger.info(f"Saved {key} to CSV.")
```

```python
# --- Main Execution ---

def main():
    logger.info("Starting NFT market analysis")

    df_raw = load_and_initial_inspect(DATA_FILE_NAME)
    if df_raw is None:
        logger.error("Data loading failed. Analysis stopped.")
        return

    df_processed = preprocess_nft_data(df_raw)
    if df_processed is None:
        logger.error("Data preprocessing failed. Analysis stopped.")
        return

    eda_results = perform_eda(df_processed)
    if not eda_results:
        logger.error("EDA failed. Analysis stopped.")
        return

    visualize_nft_data(df_processed, eda_results)

    save_results(eda_results)

    logger.info("NFT market analysis completed successfully!")

if __name__ == "__main__":
    main()
```
